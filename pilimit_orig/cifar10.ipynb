```python
import pandas as pd
import glob
import os

def load_all_pkl(folder):
    pkl_paths = glob.glob(os.path.join(folder, "*.df"))
    print(len(pkl_paths))
    list_of_dfs = []
    for path in pkl_paths:
        try:
            df = pd.read_pickle(path)
        except:
            continue
        list_of_dfs.append(df)
    return pd.concat(list_of_dfs)
```
